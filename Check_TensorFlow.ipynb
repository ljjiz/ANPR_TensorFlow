```python
import os
import requests
import tarfile

if os.name == 'posix':
    # Assuming PRETRAINED_MODEL_URL and other variables are defined
    response = requests.get(PRETRAINED_MODEL_URL)
    
    with open(os.path.join(paths['PRETRAINED_MODEL_PATH'], PRETRAINED_MODEL_NAME+'.tar.gz'), 'wb') as f:
        f.write(response.content)

    with tarfile.open(os.path.join(paths['PRETRAINED_MODEL_PATH'], PRETRAINED_MODEL_NAME+'.tar.gz'), 'r:gz') as tar:
        tar.extractall(path=paths['PRETRAINED_MODEL_PATH'])

elif os.name == 'nt':
    import wget

    # Assuming PRETRAINED_MODEL_URL and other variables are defined
    wget.download(PRETRAINED_MODEL_URL)

    # Assuming PRETRAINED_MODEL_NAME and paths dictionary are defined
    os.rename(PRETRAINED_MODEL_NAME+'.tar.gz', os.path.join(paths['PRETRAINED_MODEL_PATH'], PRETRAINED_MODEL_NAME+'.tar.gz'))

    with tarfile.open(os.path.join(paths['PRETRAINED_MODEL_PATH'], PRETRAINED_MODEL_NAME+'.tar.gz'), 'r:gz') as tar:
        tar.extractall(path=paths['PRETRAINED_MODEL_PATH'])
```
